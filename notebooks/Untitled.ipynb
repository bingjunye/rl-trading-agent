```python
import urllib.request
import pandas as pd
import json



# 用 JSON 格式抓，看看完整的錯誤訊息
url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=SPY&outputsize=full&apikey={API_KEY}"
urllib.request.urlretrieve(url, "../data/spy_response.json")

with open("../data/spy_response.json") as f:
    data = json.load(f)

print(list(data.keys()))
print(data.get("Information", data.get("Note", "no message")))
```
